```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(42)
```

```python
num_samples=1000
```

```python
square_footage=np.random.randint(800,3500,size=num_samples)
```

```python
bedrooms=np.random.randint(1,6,size=num_samples)
```

```python
bathrooms=np.random.uniform(1,4,size=num_samples)
```

```python
random_noise=np.random.normal(0,20000,size=num_samples)
```

```python
prices=100*square_footage+20000*bedrooms+15000*bathrooms+random_noise
```

```python
data=pd.DataFrame({'square footage':square_footage,'bedrooms':bedrooms,'bathrooms':bathrooms,'price':prices})
```

```python
data.to_csv('house_prices_dataset.csv',index=True)
```

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv('house_prices_dataset.csv')
```

```python
data
```

```python
x=data[['square footage', 'bedrooms', 'bathrooms',]]
```

```python
y=data['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error (y_test,y_pred)
```

```python
print("Mean Squared Error:", mse)
```

```python
example_house = [[1700 , 3, 2 ]]
```

```python
predicted_price = model.predict(example_house)
```

```python
print("predicted price:", predicted_price[0])
```
